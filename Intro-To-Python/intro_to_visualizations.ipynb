```python
x =  np.array([10, 15, 20, 25, 30, 35, 40])
y =  np.array([25, 36, 42, 55, 62, 56, 49])
font1 = {'family' : 'serif', 'color': 'green', 'size': 20}

plt.plot(x, y)
plt.xlabel('Average Hours')
plt.ylabel('Average Score')
plt.title('Line Chart', fontdict =font1, loc ='left')
plt.show()
```

```python
plt.plot(x* -1.5, y* -1.5)
plt.xlabel('Average Hours')
plt.ylabel('Average Score')
plt.title('Line Chart', fontdict =font1, loc ='left')
plt.grid(axis = 'y')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

data = sns.load_dataset('flights')
data.head()
```

```python
sns.scatterplot(data = data, x='month', y='passengers', hue='year')

plt.title('Scatter plot: Number of passengers each month')
plt.xlabel('Month')
plt.ylabel('Number of passenger')
plt.show()
```

```python
sns.boxplot(data=data,  x='month', y='passengers')
plt.xlabel('Month')
plt.ylabel('Number of passenger')
plt.grid()
plt.show()
```

```python
sns.violinplot(data=data,  x='month', y='passengers')
plt.xlabel('Month')
plt.ylabel('Number of passenger')
plt.grid()
plt.show()
```

```python
pivot_data = data.pivot_table(index='month', columns='year', values='passengers')
pivot_data.plot(kind='bar', stacked=True)
plt.show()
```

```python
sns.jointplot(data=data, x='month', y='passengers', kind='scatter')
plt.title('Joint Plot')
plt.show()
```

```python
sns.countplot(data=data, x='month')
plt.xlabel('passengers')
plt.ylabel('Count')
plt.show()
```
